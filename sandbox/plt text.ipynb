```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
df = sns.load_dataset('diamonds')

# KDE plot
sns.kdeplot(data=df, x='carat', bw_adjust=0.5)

# Add text annotation
plt.text(1.5, 2.0, 'This is a test', ha='center', va='center', fontsize=14)

# Show plot
plt.xlabel("Carat")
plt.ylabel("Density")
plt.title("KDE Plot with Text Annotation")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07', '2023-01-10'],
    'value': [5, 3, 6, 2, 7]
}

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])  # Convert to datetime

# KDE plot
sns.kdeplot(df['date'], bw_adjust=0.5)

# Add text annotation
plt.text(pd.Timestamp('2023-01-06'), 0.05, 'Important Date', ha='center', fontsize=12)

# Customize plot
plt.xlabel("Date")
plt.ylabel("Density")
plt.title("KDE Plot with Date on X-axis and Text Annotation")
plt.show()
```
