```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

# Sample text input
text = "Sample text input for generating a WordCloud"

# Define a custom colormap
custom_cmap = plt.get_cmap("rainbow")

# Function to apply the colormap to the WordCloud with proper integer conversion
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    # Get an RGB tuple from the colormap and convert to 255 scale
    color = custom_cmap(np.random.rand())[:3]
    return tuple(int(c * 255) for c in color)

# Create a WordCloud with the custom color function
wordcloud = WordCloud(color_func=color_func).generate(text)

# Display the WordCloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
