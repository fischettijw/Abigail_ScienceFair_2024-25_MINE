```python
import pandas as pd

# Sample dataframe with a datetime column
data = {'datetime': ['2024-09-26 08:30:00', '2024-09-26 19:45:00', '2024-09-26 14:00:00']}
df = pd.DataFrame(data)

# Convert the 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Function to classify as 'AM' or 'PM' based on the time
def classify_time(row):
    am_start = 6
    if am_start <= row.hour < (am_start + 12):
        return 'AM'
    else:
        return 'PM'

# Apply the function to create the new column
df['AM_PM'] = df['datetime'].apply(classify_time)

print(df)
```

```python
import pandas as pd

# Sample dataframe with a datetime column
data = {'datetime': ['2024-09-26 08:30:00', '2024-09-26 19:45:00', '2024-09-26 14:00:00']}
df = pd.DataFrame(data)

# Convert the 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Use a lambda expression to classify as 'AM' or 'PM'
df['AM_PM'] = df['datetime'].apply(lambda hr: 'AM' if 6 <= hr.hour < 18 else 'PM')

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame with a 'date' column
data = {'date': ['2024-01-15', '2024-02-20', '2024-03-10', '2024-04-05', 
                 '2024-05-30', '2024-06-18', '2024-07-12', '2024-08-25',
                 '2024-09-16', '2024-10-04', '2024-11-13', '2024-12-29'],
        'value': [5, 6, 7, 8, 2, 3, 4, 5, 6, 7, 8, 9]}

df = pd.DataFrame(data)

# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract month names
df['month'] = df['date'].dt.strftime('%b')

# Group by month and get sum or count of 'value'
df_grouped = df.groupby('month')['value'].sum()

# Ensure the months appear in correct order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_grouped = df_grouped.reindex(month_order)

# Plot the bar chart
df_grouped.plot(kind='bar')

# Set labels and title
plt.xlabel('Month')
plt.ylabel('Value')
plt.title('Values per Month')

plt.plot(2, 8, 'ro')  # 'ro' for red circle marker

# Display the plot
plt.show()
```

```python
# https://chatgpt.com/share/66f7136b-4494-8011-ab18-13cc6fe27b3f

import pandas as pd
import matplotlib.pyplot as plt

# Sample data with months
data = {
    'rain': ['Y', 'N', 'Y', 'Y', 'N', 'N', 'Y', 'N', 'Y', 'Y', 'Y', 'N'],
    'rain_AM': ['Y', 'N', 'Y', 'N', 'N', 'Y', 'Y', 'N', 'Y', 'N', 'N', 'Y'],
    'rain_PM': ['N', 'Y', 'Y', 'Y', 'N', 'N', 'Y', 'N', 'Y', 'Y', 'N', 'Y'],
    'month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
              'Oct', 'Nov', 'Dec']  # Months
}

# Create DataFrame
df = pd.DataFrame(data)
print(df.head())


# Define correct order of months
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep',
               'Oct', 'Nov', 'Dec']

# Convert 'month' column to categorical with correct order
df['month'] = pd.Categorical(df['month'], categories=month_order, ordered=True)

# Group by month and calculate rain frequency
rain_am_count = df.groupby('month')['rain_AM'].apply(lambda x: (x == 'Y').sum())
rain_pm_count = df.groupby('month')['rain_PM'].apply(lambda x: (x == 'Y').sum())

# Combine the counts into a DataFrame
rain_freq = pd.DataFrame({
    'AM Rain': rain_am_count,
    'PM Rain': rain_pm_count
})

# Create a stacked bar plot
rain_freq.plot(kind='bar', stacked=True)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Frequency of Rain')
plt.title('Stacked Bar Chart of AM and PM Rain Frequency by Month')
plt.xticks(rotation=45)

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Sample data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [35, 25, 25, 15]

# Create figure and 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Generate angles and heights for bars
theta = np.linspace(0, 2 * np.pi, len(sizes), endpoint=False)
width = 2 * np.pi / len(sizes)
heights = sizes

# Create 3D bars
ax.bar(theta, heights, width=width, zs=0, bottom=0, color=plt.cm.Paired(np.arange(len(sizes))))

# Set labels and title
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_title('3D Pie Chart')

# Add a legend
plt.legend(labels, loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

plt.style.use("dark_background")
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

def plot_wedge(ax, start=0, stop=2*np.pi, r=10, h=1, cen_x=0, cen_y=0, cen_z=0, c=None):
    l = 1000

    t = np.linspace(start, stop, l)
    x = np.concatenate([[cen_x], r*np.cos(t)+cen_x, [cen_x]])
    zeros = np.zeros(len(x))
    x_sides = np.stack([x, x])
    x_top = np.stack([x, zeros + cen_x])
    y = np.concatenate([[cen_y], r*np.sin(t)+cen_y, [cen_y]])
    y_sides = np.stack([y, y])
    y_top = np.stack([y, zeros+cen_y])
    z_sides = np.array([[cen_z], [cen_z+h]])
    z_top = np.array([[cen_z+h]])

    ax.plot_surface(x_top, y_top, z_top, color=c)
    ax.plot_surface(x_sides, y_sides, z_sides, color=c)

plot_wedge(ax, np.pi/2, 10/5*np.pi, c="#79b81a")
plot_wedge(ax, np.pi/3, 14/10*np.pi, r=9, cen_x=1, cen_y=1, cen_z=1, c="#15b269")
plot_wedge(ax, np.pi/3, 12/10*np.pi, r=8, cen_x=4, cen_y=4, cen_z=2, c="#0688a6")
plot_wedge(ax, np.pi/6, 8/10*np.pi, r=5, cen_x=1, cen_y=1, cen_z=3, c="#d92161")
plot_wedge(ax, np.pi/9, 5/10*np.pi, r=3, cen_x=1, cen_y=1, cen_z=4, c="#f65c5a")
ax.set_xlim(-15, 15)
ax.set_ylim(-15, 15)
ax.set_zlim(0, 10)
ax.set_axis_off() 
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, layout="constrained")

# Matplotlib marker symbol
axs[0, 0].scatter(x, y, s=80, c=z, marker=">")
axs[0, 0].set_title("marker='>'")

# marker from TeX: passing a TeX symbol name enclosed in $-signs
axs[0, 1].scatter(x, y, s=80, c=z, marker=r"$\clubsuit$")
axs[0, 1].set_title(r"marker=r'\$\clubsuit\$'")

# marker from path: passing a custom path of N vertices as a (N, 2) array-like
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0, 2].scatter(x, y, s=80, c=z, marker=verts)
axs[0, 2].set_title("marker=verts")

# regular pentagon marker
axs[1, 0].scatter(x, y, s=80, c=z, marker=(5, 0))
axs[1, 0].set_title("marker=(5, 0)")

# regular 5-pointed star marker
axs[1, 1].scatter(x, y, s=80, c=z, marker=(5, 1))
axs[1, 1].set_title("marker=(5, 1)")

# regular 5-pointed asterisk marker
axs[1, 2].scatter(x, y, s=80, c=z, marker=(5, 2))
axs[1, 2].set_title("marker=(5, 2)")

plt.show()

# https://matplotlib.org/stable/api/markers_api.html           MARKERS
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/scatter_star_poly.html       MARKER EXAMPLES
```
