```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data: Heights of students
heights = [150, 155, 160, 162, 165, 167, 170, 172, 175, 178, 180, 182, 185, 190]

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# First plot (non-cumulative)
sns.kdeplot(heights, fill=True, gridsize=100, cumulative=False, ax=axes[0])
axes[0].set_xlabel('Height (cm)')
axes[0].set_ylabel('Density')
axes[0].set_title('Non-Cumulative KDE')

# Second plot (cumulative)
sns.kdeplot(heights, fill=True, gridsize=100, cumulative=True, ax=axes[1])
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Density')
axes[1].set_title('Cumulative KDE')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```
